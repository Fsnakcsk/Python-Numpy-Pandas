# t-검정(차이검정) 정규분포에 따르냐

```python
# t-검정
from scipy import stats

# 회귀분석                     보고서 출력 , 회귀분석할 함수
from statsmodels.formula.api import ols,  glm
```

```python
#품질만 따로 뽑아온
red_wine_quality = wine.loc[wine.type == 'red', 'quality']
white_wine_quality = wine.loc[wine.type == 'white', 'quality']
```

```python
# 독립 T 검정 (두개의 객체 서로 길이 다를 때)
# stats.ttest_ind(1, 2)                  

print('red   :',np.var(red_wine_quality))
print('white :',np.var(white_wine_quality))
#print(ttest_inds(red_wine_quality, white_wine_quality))

# 분산(서로 떨어지는 거리의 길이)의 평균이 다르기 때문에  equal_var=False 사용함
print(stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var = False))

#-24성  = 귀목가설 기각
```

- 단일표본 T검정 : ttest-1samp(표본데이터, 귀무가설의 기대값)
- 독립표본 T검정 (정규분포에 따르다, 등분산인지 아닌지 판단 아니면 equal_var=False) : stats.ttest_inde(a, b, equal_var=True)
- 대응표본 T검정 : ttest_rel(a, b) 

- 등분산검정 : ttest_inds(a, b)
- 정규성검정 : shapiro()

```python
scales_columns = ['fixed_acidity','volatile_acidity' , 'citric_acid' ,
                  'residual_sugar' , 'chlorides' , 'free_sulfur_dioxide' , 'total_sulfur_dioxide',
                  'density' , 'pH' , 'sulphates' , 'alcohol']
```

```python
def standard_scaling(df, scale_columns) :
    for col in scale_columns :
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x : (x-series_mean) / series_std)
    return df
```

```python
sample1_df = standard_scaling(sample1, scales_columns)
sample1_df.head()
```
