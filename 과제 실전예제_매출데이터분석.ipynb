```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
file_path = "chipotle.csv"
```

```python
chipo = pd.read_csv(file_path)
print(type(chipo))
```

```python
chipo.info()
```

```python
chipo.item_price = chipo.item_price.apply(lambda x : x[1:]).astype(float)
```

```python
chipo.head()
```

```python
# 가장 많이 판매한 제품의 수량은 1개와 2개의 판매액 보기
product_count = chipo[['item_name', 'quantity','item_price']].groupby(['item_name','quantity'], 
                                                         as_index= False).agg({'item_price':sum})
product_count.head()

#인기 많은 상품은 1개만 구매 하는 경우 더 많다.
```

```python
# 제일 많이 판매한 제품 순서대로 나열
product_count = chipo[['item_name', 'quantity']].groupby(['item_name'],
                                                         as_index= False).agg({'quantity':sum}).sort_values(by='quantity',ascending=False) 
product_count.head()
```

```python
# 인기 많은 제품중에서 토핑 상황 ****
a = pd.DataFrame(chipo[chipo['choice_description'].notnull()].groupby('item_name')['quantity'].sum().sort_values(ascending=False))
a

# 인기 많은 제품은 토핑도 필수인 것같다.
```

```python
# 총 판매건수
str(chipo['order_id'].nunique()) + '개'
```

```python
# 총 몇개의 제품 판매했는지 계산
str(chipo['item_name'].nunique()) + '개'
```

```python
b = chipo[chipo['quantity'] == 1].groupby('item_name')[['item_price']].mean().sort_values(by = 'item_price', ascending = False)
b
```

```python
pd.merge(b, a, how='outer', on='item_name').reset_index().sort_values(by = ['item_price','quantity'], ascending = False)
```

```python
# 인기 많은 제품의 총판매액 
product_count = chipo[['item_name', 'item_price']].groupby('item_name', 
                                                         as_index= False).agg({'item_price':sum}).sort_values(by='item_price',ascending=False) 
product_count.head()
```

```python
# 데이터 수집한 동안의 총판매액

tot = chipo['item_price'].sum()
tot
```

```python
# 각 제품의 판맥액은 전체 판매액의 차지율

product_count['백분율'] = product_count['item_price'].apply(lambda x : str(round((x/tot)*100,2))+'%')
product_count.head(15)

# 50개의 판매 제품중에서 인기Top 4 제품은 거의 48%로 자치하고 있다.
# 대략 나머지 품중에서의 40개정도는 팔리지 않는다.
```
