## 바 차트(bar chart)
- bar
- barh
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh

```python
y = [2, 3, 1]
x = np.arange(len(y))

plt.title('Bar chart')
xlabel = ['가','나','다']
plt.bar(x, y) # 막대 그래프
plt.xticks(x, xlabel)  # 눈금
plt.yticks(sorted(y))
plt.xlabel('x축 제목')
plt.ylabel('빈도수')
plt.show()
```

```python
np.random.seed(0)
people = ['몽룡','춘향','방자','향단']
y_pos = np.arange(len(people))
data = np.random.rand(len(people)) * 10
error = np.random.rand(len(people))

plt.title('수평막대그래프')
plt.barh(y_pos, data, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x라벨')
plt.show()
```

```python
# 스템 플롯 : 바차트와 유사하지만 폭이 없는 그래프
# 주로 이산확률함수나 자기상관관계를 묘사할 때 사용

x = np.linspace(0.1, 2*np.pi, 10)
plt.title('Stem Plot')
plt.stem(x, np.cos(x), '-.')
plt.show()
```

## 파이 차트(pie chart)
- 카테고리별 값의 상대적인 비교를 해야할 떄 주로 사용한다.
- 파이 차트를 그릴 때 원의 형태를 유지할 수 있도록 plt.asix('equal') 명령어를 실행해야 한다
- https://matplotlib.org/2.0.2/examples/pie_and_polar_charts/pie_demo_features.html

```python
labels = ['맥주','소주','양주','막걸리']
sizes = [25, 45, 20, 10]
colors = ['yellowgreen','lightblue','gold','lightcoral']
explode = (0, 0.1, 0, 0)

plt.title('내가 간절히 원하는 것은 무엇') 
#                                   각영역 이름     색상             표현식         음영          각도
plt.pie(sizes, explode = explode, labels = labels, colors=colors, autopct="%.1f%%", shadow=True, startangle=-80)
plt.axis('equal')
plt.show()
```

## scatter plot : 산점도(산포도)
- 2차원 데이터 즉, 두 개의 연속형(실수)데이터 집합의 상관관계를 파악할 때 유용한 시각화 도구
- 점 하나의 위치(x , y)의 교차의 값

```python
np.random.seed(0)
x = np.random.normal(0,1, 100) #정규분포
y = np.random.normal(0,1, 100)
plt.title('Scatter Plot')
plt.scatter(x, y)
plt.show()
```

```python
# 버블차트(bubble chart) : 2차원이 아니라 3차원 또는 4차원의 산점도 표현할 때 점하느의 크기 혹은
# 색깔을 이용하여 다른 데이터값을 표시함 (s크기,c색깔)

N = 30
np.random.seed(0)
x = np.random.rand(N)


y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N)) ** 2

plt.title('Bubble Chart')
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

## imshow() : (image show 화상데이터) - 데이터를 표현하는 시각화 도구
- 이미지데이터도 행과 열을 가진 2차원 데이터이다
https://matplotlib.org/2.0.2/api/pyplot_api.html
- imshow
- nearest

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits
```

```python
x = digits.images[0]
x
```

```python
plt.title('손글씨 데이터셋(mnist digits : 0)')
#                  보간법          방식         표준 스칼라 매핑
plt.imshow(x, interpolation = 'bilinear', cmap=plt.cm.bone_r)
plt.xticks([])
plt.yticks([])
plt.grid(False) # 눈금
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65) # F - subplots_adjust
```

```python
methods = ['none', 'nearest', 'bilinear', 'bicubic', 'spline16', 
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 
           'quadric', 'catrom', 'gaussian', 'bessel', 'mitchell', 
           'sinc', 'lanczos']
#     18                행,열     도화지 사이즈
fig, axes = plt.subplots(3, 6, figsize=(14, 6),
                        subplot_kw = {'xticks':[], 'yticks':[]})

for ax, inter_method in zip(axes.flat, methods) : 
    ax.imshow(x, interpolation = inter_method, cmap=plt.cm.bone_r)
    ax.set_title(inter_method)
plt.show()
```

## Seaborn을 사용한 데이터 분포 시각화
- http://seaborn.pydata.org/
- Seaboran은 Matplotlib을 기반으로 다양한 색상과 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지
- 기본적인 시각화 기능은 Matplotlib 패키지에 의존, 통계기능은 Statsmodels 패키지에 의존

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
iris = sns.load_dataset('iris') # 붗꼭 데이터
titanic = sns.load_dataset('titanic') # 타이타닉호 데이터
flights = sns.load_dataset('flights') # 여객 운송 데이터
```

```python
iris
```

```python
# 1차원 실수 분포 플롯
# rugplot 실수 밀도 
# rugplot 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내어 실제 데이터의 위치를 보여준다.

x = iris.petal_width.values
sns.kdeplot(data=iris, x="petal_width")
sns.rugplot(x) # 데이터의 위치, 실제데이터의 위치를 작은 눈금으로 표시함
plt.ylim(0.0, 1.0)
plt.show()
```

```python
sns.displot(x, kde=True, rug=True)
plt.show()
```

```python
sns.countplot(x='class', data=titanic)
plt.title('타이타닉호 클래스별 고객수')
plt.show()
```

```python
sns.countplot(x='day', data=tips)
plt.title('요일별 팀을 준 횟수')
plt.show()
```

```python
# 다차원, data = iris, kind='kde'
sns.jointplot(x='sepal_length', y='sepal_width', data = iris)
plt.show()
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data = iris, kind='kde')
plt.show()
```

```python
# 다차원 실수형 데이터 시각화
# 삼점도 행렬 pairplot
sns.pairplot(data=iris)
plt.show()
```

```python
# hue(색상) : 카테고리 변수를 지정하여 카테고리에 따라 색상을 다르게 지정함
# hue(색상): 점주형 데이터 많이 사용한다
sns.pairplot(data=iris, hue = 'species', markers=['o','s','D'])
plt.show()
```

```python
titanic_size = titanic.pivot_table(index = 'class', columns = 'sex', aggfunc='size')
titanic_size
```

```python
# 2차원 카테고리형 데이터(범주형) : heatmap
# heatmap : http://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap
sns.heatmap(titanic_size, cmap = 'crest', annot=True, fmt='d')
plt.show()
```
